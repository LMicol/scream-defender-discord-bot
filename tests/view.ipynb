```python
import matplotlib.pyplot as plt
import pickle as pkl
import numpy as np
```

```python
with open('audio.pkl', 'rb') as f:
    audio = pkl.load(f)
```

```python
sample_rate = 148000
```

```python
if audio.ndim == 2:
    audio_data = audio.mean(axis=1)

n = len(audio_data)
freq = np.fft.fftfreq(n, d=1/sample_rate)
spectrum = np.fft.fft(audio_data)

# Get the positive half of the spectrum and corresponding frequencies
positive_freqs = freq[:n//2]
positive_spectrum = np.abs(spectrum[:n//2])

# Plot the spectrum
plt.figure(figsize=(10, 6))
plt.plot(positive_freqs, positive_spectrum)
plt.title('Audio Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
# Create a time array corresponding to the audio samples
n_samples = len(audio_data)
time = np.linspace(0, n_samples / sample_rate, num=n_samples)

# Plot the waveform
plt.figure(figsize=(10, 6))
plt.plot(time, audio_data)
plt.title('Audio Waveform')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
